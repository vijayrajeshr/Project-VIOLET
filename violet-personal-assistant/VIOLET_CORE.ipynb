```python
# Phase 1: The VIOLET HUD (GUI)
import ipywidgets as widgets
from IPython.display import display, clear_output
import psutil
import threading
import time

status_dot = widgets.Label(value="🔴")
status_text = widgets.Label(value="VIOLET: OFFLINE", layout={'width': '150px'})
toggle_btn = widgets.ToggleButton(
    value=False, 
    description='ENABLE BEAST MODE', 
    button_style='danger',
    tooltip='Activate VIOLET System Control',
    icon='power-off',
    layout={'width': '200px'}
)
cpu_bar = widgets.IntProgress(
    value=0, min=0, max=100, 
    description='CPU LOAD:', 
    bar_style='info', 
    orientation='horizontal',
    layout={'width': '100%'}
)
task_log = widgets.Textarea(
    value='', 
    placeholder='System events will appear here...',
    description='LOG:',
    disabled=True,
    layout={'height': '150px', 'width': '100%'}
)

def log_event(msg):
    t = time.strftime("%H:%M:%S")
    task_log.value = f"[{t}] > {msg}\n" + task_log.value

def on_toggle(change):
    if change['new']:
        toggle_btn.description = 'VIOLET ACTIVE'
        toggle_btn.button_style = 'success'
        status_dot.value = "🟢"
        status_text.value = "VIOLET: ONLINE"
        log_event("Beast Mode Engaged. System privileges granted.")
    else:
        toggle_btn.description = 'Initiate Project-Violet'
        toggle_btn.button_style = 'danger'
        status_dot.value = "🔴"
        status_text.value = "VIOLET: OFFLINE"
        log_event("VIOLET entering standby...")

toggle_btn.observe(on_toggle, 'value')

def update_stats():
    while True:
        cpu_bar.value = int(psutil.cpu_percent())
        time.sleep(1)

threading.Thread(target=update_stats, daemon=True).start()

header = widgets.HBox([status_dot, status_text, toggle_btn], layout={'justify_content': 'space-between', 'padding': '10px'})
dashboard = widgets.VBox([
    widgets.HTML("<b>VIOLET CORE v1.0</b>"),
    header, 
    cpu_bar, 
    task_log
], layout={'border': '1px solid #444', 'padding': '15px', 'width': '450px', 'background': '#1e1e1e'})

display(dashboard)
```
